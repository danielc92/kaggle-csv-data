### Imports

```python
import pandas
import os
import requests
import bs4
```

### Read in Table

```python
data = pandas.read_excel('./table.xlsx')
```

```python
data
```

```python
data_path = './html_descs/'
base_url = 'https://vincentarelbundock.github.io/Rdatasets/'
```

### Pull in HTML 

```python
for index in data.index:
    print('processing index: {}/{}'.format(str(index), data.shape[0]))
    name = data.loc[index, 'Unique_Name']
    file_name = name + '.html'
    suffix = data.loc[index, 'Doc']
    url = base_url + suffix
    r = requests.get(url)
    s = bs4.BeautifulSoup(r.content)
    with open(os.path.join(data_path, file_name), 'w') as f:
        f.write(str(s))
        f.close()
```

### Get Names

```python
names = list(data['Unique_Name'])
```

```python
names
```

```python
def return_html(path):
    with open(path, 'r') as file:
        data = file.read()
        file.close()
    
    return data
```

```python
html_data = [return_html('./html_descs/{}.html'.format(file)) for file in names]
```

### Clean whitespace

```python
def remove_whitespace(test):

    while '<h3> ' in test:
        test = test.replace('<h3> ', '<h3>')

    while ' </h3>' in test:
        test = test.replace(' </h3>', '</h3>')

    while '<h2> ' in test:
        test = test.replace('<h2> ', '<h2>')

    while ' </h2>' in test:
        test = test.replace(' </h2>', '</h2>')
        
    return test
```

```python
html_data_clean = [remove_whitespace(item) for item in html_data]
```

```python
len(html_data_clean)
```

<h2> --> FIND TITLE
'<h3>Usage</h3>' --> FIND <pre>
'<h3>Description</h3>' --> FIND <p>
'<h3>Source</h3>' --> FIND <p>

```python
for ii, item in enumerate(html_data_clean):    
    # GET TITLE
    soup = bs4.BeautifulSoup(item)
    title = soup.find('h2').text.strip()
    
    # GET SOURCE
    sub_index = item.find('<h3>Source</h3>')
    substring = item[sub_index:].replace('<h3>Source</h3>', '')
    another_index = substring.find('<h3>')
    substring = substring[:another_index]
    soup = bs4.BeautifulSoup(substring)
    ps = soup.find_all('p')
    temp = []
    for i, p in enumerate(ps):
        if i==0 or i ==1:
            temp.append(p.text.strip())
    sources = ' '.join(temp)
    
    # GET SOURCE
    sub_index = item.find('<h3>Description</h3>')
    substring = item[sub_index:].replace('<h3>Description</h3>', '')
    another_index = substring.find('<h3>')
    substring = substring[:another_index]
    print(substring)
    soup = bs4.BeautifulSoup(substring)
    ps = soup.find_all('p')
    temp = []
    for i, p in enumerate(ps):
        if i==0 or i ==1:
            temp.append(p.text.strip())
    descs = ' '.join(temp)
    
    data.loc[ii, 'trans_title'] = title
    data.loc[ii, 'trans_description'] = descs
    data.loc[ii, 'trans_sources'] = sources
    
```
